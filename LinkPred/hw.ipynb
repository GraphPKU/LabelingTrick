```python
import numpy as np
```

```python
w1=[(0,1),(1,1),(1,3),(4,3),(7,0),(7,3),(8,1),(8,2),(8,3),(9,0)]
w2=[(0,5),(0,6),(1,5),(1,8),(2,6),(2,7),(2,8),(3,6),(4,6),(4,8)]
w3=[(4,5),(6,4),(6,8),(7,4),(8,4),(9,4),(9,6),(9,7),(4,9),(6,7)]
w1,w2,w3=np.array(w1),np.array(w2),np.array(w3)
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
vor = Voronoi(np.concatenate((w1, w2, w3), axis=0))
```

```python
import matplotlib.pyplot as plt
```

```python
fig = voronoi_plot_2d(vor)
```

```python
vor.point_region = np.concatenate((np.zeros_like(w1[:, 0]), np.ones_like(w2[:, 0]), 2*np.ones_like(w3[:, 0])))
```
