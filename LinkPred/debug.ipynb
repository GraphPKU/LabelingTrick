```python
import scipy.sparse as spsp
```

```python
def data2graph(data):
    node_id = data.node_id
    ei = node_id[data.edge_index]
    ea = data.edge_weight
    return spsp.csr_matrix((ea.numpy(), (ei[0].numpy(), ei[1].numpy())))

def data2z(data):
    node_id = data.node_id
    ei = node_id[data.edge_index]
    ea = data.edge_weight
    return spsp.csr_matrix((ea.numpy(), (ei[0].numpy(), ei[1].numpy())))
```

```python
def collatepos(pos: torch.Tensor):
    '''
    pos (batch_size, 3), the third column is subgraph size
    '''
    tpos = pos[:, :2].clone()
    tpos[1:] += torch.cumsum(pos[:-1, 2], dim=0).unsqueeze(-1)
    return tpos
```
